W6A9

```python
def credict():
    d = {}
    nhap = input().split()
    for i in nhap:
        key, value = i.split(":")
        d[key] = value
    return d

a = credict()
b = credict()

for i in a:
    if i in b:
        b[i] = int(a[i]) + int(b[i])
    else:
        b[i] = a[i]

print(dict(sorted(b.items())))
```

```python
a = list(map(int, input().split()))
k = int(input())

kq = []

for i in a[:]:
    a.remove(i)
    if (k-i) in a:
        s = []
        s.append(i)
        s.append(k-i)
        kq.append(tuple(s))
print(list(dict.fromkeys(kq)))
```
