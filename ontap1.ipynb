```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    if car_type == 4:
        gia1 = 11000
        gia2 = 13000
        gia3 = 10500
    elif car_type == 7:
        gia1 = 12000
        gia2 = 14000
        gia3 = 12000
    
    if km <= 0.8:
        tien = gia1
    elif km <= 20:
        tien = gia1 + (km-0.8)*gia2
    else:
        tien = gia1 + (20-0.8)*gia2 + (km-20)*gia3
    
    phicho = (wait_minutes//5)*2000

    tongtien = tien + phicho

    night_hour = [22, 23, 0, 1, 2, 3, 4, 5]

    if start_hour in night_hour:
        tongtien *= 1.2
    elif km > 30 and start_hour not in night_hour:
        tongtien *= 0.9
    
    if tongtien < 50000 and km > 5:
        tongtien = 50000
    return int(tongtien)

car_type, start_hour, km, wait_minutes = map(int, input().split())
print(smart_taxi_fare(car_type, start_hour, km, wait_minutes))
```

```python
def  quadratic_roots_count(a, b, c):
    delta = b**2 - 4*a*c
    nghiem = []
    if a == 0 and b != 0:
        x = -c/b
        nghiem.append(x)
    elif a == 0 and b == 0:
        if c == 0:
            return "Vô số nghiệm"
        else:
            return "Vô nghiệm"
    elif a != 0:
        if delta > 0:
            x1 = (-b + delta**0.5) / (2*a)
            x2 = (-b - delta**0.5) / (2*a)
            nghiem.append(x1)
            nghiem.append(x2)
        elif delta == 0:
            x = -b / (2*a)
            nghiem.append(x)
    nghiem.sort()
    return nghiem

a, b, c = map(int, input().split())
print(quadratic_roots_count(a, b, c))
```

```python
import math

def tichvohuong(a, b, c):
    return (a[0] - b[0]) * (c[0] - b[0]) + (a[1] - b[1]) * (c[1] - b[1])

def dodai(x, y):
    return ((x[0] - y[0])**2 + (x[1] - y[1])**2)**0.5

def goc(a, b, c):
    ba = dodai(a, b)
    bc = dodai(c, b)
    radian = math.acos(tichvohuong(a, b, c) / (ba * bc))
    do = math.degrees(radian)
    return do

a = tuple(map(int, input().split()))
b = tuple(map(int, input().split()))
c = tuple(map(int, input().split()))
print(goc(a, b, c))
```

```python
def point_in_quadrilateral(quad, p):

    def tichcohuong(a, b, c):
        return (b[0] - a[0]) * (c[1] - a[1]) - (b[1] - a[1]) * (c[0] - a[0])
    
    v1 = tichcohuong(quad[0], quad[1], p)
    v2 = tichcohuong(quad[1], quad[2], p)
    v3 = tichcohuong(quad[2], quad[3], p)
    v4 = tichcohuong(quad[3], quad[0], p)

    return (v1 >= 0 and v2 >= 0 and v3 >= 0 and v4 >= 0) or (v1 <= 0 and v2 <= 0 and v3 <= 0 and v4 <= 0)

quad = [tuple(map(int, input().split())) for _ in range(4)]

p = tuple(map(int, input().split()))

print(point_in_quadrilateral(quad, p))
```
