```python
n = int(input())

a = n
sochan = 0
sole = 0

while True:
    if (n%10)%2 == 0:
        sochan += 1
    elif (n%10)%2 != 0:
        sole += 1
    n //= 10
    if n == 0:
        break
print(f"so {a} co {sochan} so chan, {sole} so le")
```

W4A29

```python
n = input()

new = ""
for i in n:
    if 97<= ord(i)<= 122:
        i = chr(ord(i)-32)
    elif 65<= ord(i)<= 90:
        i = chr(ord(i)+32)
    new += i
print(n, new)
```

```python
#0-9:48-57
n = input()

kytuhoa = 0
kytuthuong = 0
kytuso = 0
for i in n:
    if 97<= ord(i)<= 122:
        kytuthuong += 1
    elif 65<= ord(i)<= 90:
        kytuhoa += 1
    elif 48<= ord(i)<= 57:
        kytuso += 1
print(n)
print("hoa", kytuhoa)
print("thuong", kytuthuong)
print("so", kytuso)
```
