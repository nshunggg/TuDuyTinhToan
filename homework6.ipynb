```python
n, m = map(int, input().split())

matrix = []

for i in range(n):
    row = []
    for j in range(m):
        a = int(input())
        row.append(a)
    matrix.append(row)

for i in range(n):
    for j in range(m):
        print(f"{matrix[i][j]:>4}", end="")
    print()
```

```python
n = int(input())

matrix = []

for i in range(n):
    row = []
    for j in range(n):
        a = int(input())
        row.append(a)
    matrix.append(row)

print("Cheo chinh")
for i in range(n):
    for j in range(n):
        if i == j:
            print(matrix[i][j], end=" ")
        else:
            print(" ", end=" ")
    print()

print("Cheo phu")
for i in range(n):
    for j in range(n):
        if i == n - j - 1:
            print(matrix[i][j], end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
n, m, k = map(int, input().split())

matrix = []

for i in range(n):
    row = []
    for j in range(m):
        a = int(input())
        row.append(a)
    matrix.append(row)

tonghangk = 0
for x in matrix[k-1]:
    tonghangk += x

tongcotk = 0
for i in range(n):
    for j in range(m):
        if j == k - 1:
            tongcotk += (matrix[i][j])

print(tongcotk, tonghangk)
```

```python
def vitri(s, k):
    if k in s:
        return s.index(k) + 1
    else:
        return -1
s = list(map(int, input().split()))
k = int(input())

print(vitri(s, k))
```

```python
s = input().split()

d = {}
for i in range(len(s)):
    key = s[i]
    value = i
    if key in d:
        d[key].append(value)
    else:
        d[key] = [value]
for a in d:
    d[a] = tuple(d[a])
print(d)
```

```python
def remove_in_list(s, x):
    for a in s[:]:
        if a < x:
            s.remove(a)
    return s
s = list(map(int, input().split()))
x = int(input())
print(remove_in_list(s, x))
```

W6A24

```python
def dayso0(s):
    count0 = 0
    for i in s:
        if i == 0:
            count0 += 1
    for _ in range(count0):
        s.remove(0)
    for _ in range(count0):
        s.insert(0, 0)
    return s
s = list(map(int, input().split()))
print(dayso0(s))
```
