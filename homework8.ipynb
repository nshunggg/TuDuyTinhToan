```python
class Shape:
    def area(self):
        pass
class Circle(Shape):
    def __init__(self, r):
        self.r = r
    def area(self):
        return 3.14 * self.r ** 2
class Rectangle(Shape):
    def __init__(self, w, h):
        self.w = w
        self.h = h
    def area(self):
        return self.w * self.h
shape_type = input()
if shape_type == "RECTANGLE":
    w, h = map(float, input().split())
    shape = Rectangle(w, h)
    print(shape_type + ": " + f"{shape.area():.2f}")
elif shape_type == "CIRCLE":
    r = float(input())
    shape = Circle(r)
    print(shape_type + ": " + f"{shape.area():.2f}")
```

```python
class Vehicle:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed
    def info(self):
        return f"Brand: {self.brand}, Speed: {self.speed} km/h"
class Car(Vehicle):
    def __init__(self, brand, speed, seats):
        super().__init__(brand, speed)
        self.seats = seats
    def info(self):
        return super().info() + f", Seats: {self.seats}"
class_type = input()
if class_type == "CAR":
    brand = input()
    speed = float(input())
    seats = int(input())
    car = Car(brand, speed, seats)
    print(car.info())
elif class_type == "VEHICLE":
    brand = input()
    speed = float(input())
    vehicle = Vehicle(brand, speed)
    print(vehicle.info())
```

```python
class Polygon:
    def perimeter(self):
        pass
class Triangle(Polygon):
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    def perimeter(self):
        return self.a + self.b + self.c
a, b, c = map(float, input().split())
triangle = Triangle(a, b, c)
if a + b > c and a + c > b and b + c > a:
    print("Perimeter: " + f"{round(triangle.perimeter(), 2)}")
else:
    print("Invalid triangle")
```

```python
class Account:
    def __init__(self, balance):
        self.balance = balance
    def deposit(self, amount):
        self.balance += amount
    def get_balance(self):
        return self.balance
class SavingsAccount(Account):
    def add_interest(self, rate):
        x = self.balance * rate / 100
        self.deposit(x)
b, r = map(float, input().split())
sa = SavingsAccount(b)
sa.add_interest(r)
print("Balance: " + f"{sa.get_balance():.2f}")
```

```python
class Printer:
    def __init__(self, value):
        self.value = value
    def print_value(self):
        return self.value
class BinaryPrinter(Printer):
    def print_value(self):
        bin_value = []
        while self.value != 1:
            bin_value.insert(0, self.value % 2)
            self.value //= 2
        bin_value.insert(0, 1)
        return "".join(map(str, bin_value))
value = int(input())
p = Printer(value)
bp = BinaryPrinter(value)
print("Decimal: " + str(p.print_value()))
print("Binary: " + bp.print_value())
```

```python
class Matrix2x2:
    def __init__(self, a11, a12, a21, a22):
        self.a11 = a11
        self.a12 = a12
        self.a21 = a21
        self.a22 = a22
    def trace(self):
        return self.a11 + self.a22
    def determinant(self):
        return self.a11 * self.a22 - self.a12 * self.a21
a11, a12, a21, a22 = map(float, input().split())
matrix = Matrix2x2(a11, a12, a21, a22)
print("Trace: " + f"{matrix.trace():.2f}")
print("Determinant: " + f"{matrix.determinant():.2f}")
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price # gia ban
    def final_price(self):
        return self.price
class DiscountProduct(Product):
    def __init__(self, name, price, percent):
        super().__init__(name, price)
        self.percent = percent # phan tram giam gia
    def final_price(self):
        self.price -= self.price * self.percent / 100
        return self.price
name = input()
price, percent = map(float, input().split())
dis_product = DiscountProduct(name, price, percent)
print("Final Price: " + f"{dis_product.final_price():.2f}")
```

```python
class Counter:
    count = 0
    def inc(self):
        Counter.count += 1
    def dec(self):
        Counter.count -= 1
    def value(self):
        return Counter.count
counter = Counter()
n = int(input())
thaotac = list(map(str, input().split()))
for i in thaotac:
    if i == "INC":
        counter.inc()
    elif i == "DEC":
        counter.dec()
print("Counter Value: " + str(counter.value()))
```

```python
class Time:
    def __init__(self, h, m, s):
        self.h = h
        self.m = m
        self.s = s
    def to_seconds(self):
        return self.h * 3600 + self.m * 60 + self.s
h, m, s = map(int, input().split())
time = Time(h, m, s)
print("Total Seconds: " + str(time.to_seconds()))
```

```python
class Stack:
    def __init__(self):
        self.stack = []
    def push(self, x):
        self.stack.append(x)
    def pop(self):
        if len(self.stack) > 0:
            self.stack.pop()
    def top(self):
        if len(self.stack) > 0:
            return self.stack[-1]
stack = Stack()
n = int(input())
for _ in range(n):
    option = input()
    if option == "PUSH x":
        x = input()
        stack.push(x)
    elif option == "POP":
        stack.pop()
    elif option == "TOP":
        print("Top: " + str(stack.top()))
```
