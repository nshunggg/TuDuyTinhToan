```python
def sortday(arr):
    l = []
    r = []
    for i in arr:
        if i % 2 == 0:
            l.append(i)
        else:
            r.append(i)
    l.sort()
    r.sort()
    return l + r[::-1]
arr = list(map(int, input().split()))
print(*sortday(arr))
```

```python
def merge(arr1, arr2):
    if len(arr1) == 0 or len(arr2) == 0:
        return []
    i = j = 0
    res = []
    tang = True
    for m in range(1, len(arr1)):
        if arr1[m] < arr1[m-1]:
            tang = False
            break
    giam = True
    for n in range(1, len(arr1)):
        if arr1[n] > arr1[n-1]:
            giam = False
            break
    while i < len(arr1) and j < len(arr2):
        if tang:
            if arr1[i] < arr2[j]:
                res.append(arr1[i])
                i += 1
            else:
                res.append(arr2[j])
                j += 1
        elif giam:
            if arr1[i] > arr2[j]:
                res.append(arr1[i])
                i += 1
            else:
                res.append(arr2[j])
                j += 1
    res.extend(arr1[i:])
    res.extend(arr2[j:])
    return res
try:
    arr1 = list(map(int, input().split()))
except EOFError:
    arr1 = []
try:
    arr2 = list(map(int, input().split()))
except EOFError:
    arr2 = []
print(*merge(arr1, arr2))
```

```python
def tinhngay(n, t):
    match t:
        case 1|3|5|7|8|10|12:
            maxday = 31
        case 4|6|9|11:
            maxday = 30
        case 2:
            maxday = 29
        case _:
            maxday = -1
    if n < 1 or n > maxday or maxday == -1:
        return "INVALID"
    tong = n
    if t == 1:
        return tong
    for i in range(1, t):
        match i:
            case 1|3|5|7|8|10|12:
                tong += 31
            case 4|6|9|11:
                tong += 30
            case 2:
                tong += 29
    return tong
n, t = map(int, input().split())
print(tinhngay(n, t))
```

```python
def numbermax(n, k):
    max_num = n[0]
    arr = []
    pos = 0
    for i in range(len(n)-k+1):
        if n[i] > max_num:
            max_num = n[i]
            pos = i
    arr.append(max_num)
    k -= 1
    if k == 0:
        return arr
    return arr + numbermax(n[pos+1:], k)
n = input()
k = int(input())
print("".join(numbermax(n, k)))
```

```python
def is_valid(s):
    n = []
    i, j = 0, 1
    s += " "
    while i <= j and j < len(s):
        if 48 <= ord(s[j])<= 57:
            j += 1
        if 48 <= ord(s[i])<= 57 and (ord(s[j]) < 48 or ord(s[j]) > 57):
            n.append(int(s[i:j]))
            i = j
        if ord(s[j]) < 48 or ord(s[j]) > 57:
            i += 1
            j += 1
    tong = 0
    for num in n:
        tong += num
    if tong % len(n) == 0:
        tb = tong // len(n)
    else:
        tb = round(tong/(len(n)), 2)
    return [tong, tb]
s = input()
print(*is_valid(s))
        
```

```python
def longest(n):
    if n <= 4:
        return "Khong co day thoa man"
    arr = []
    num = []
    nums = []
    for i in range(2, n):
        check = False
        for j in range(2, i):
            if i % j == 0:
                check = True
        if check:
            arr.append(i)
    if len(arr) == 1:
        return arr[0]
    for i in range(len(arr)-1):
        num.append(arr[i])
        if arr[i+1] - arr[i] != 1:
            nums.append(num)
            num = []
        if i + 1 == len(arr)-1:
            num.append(arr[i+1])
            nums.append(num)
    max_len = 0
    for i in nums:
        if len(i) > max_len:
            max_len = len(i)
            res = i
    return ", ".join(map(str, res))
n = int(input())
print(longest(n))
        
```

```python
def dayconklt(arr):
    num = []
    for i in arr:
        l, r = 0, len(num)-1
        pos = len(num)
        if len(num) > 0:
            if i != num[-1]:
                while l <= r:
                    m = l + (r-l)//2
                    if num[m] >= i:
                        pos = m
                        r = m - 1
                    else:
                        l = m + 1
        if pos == len(num):
            num.append(i)
        else:
            num[pos] = i
    return num
arr = list(map(int, input().split()))
print(*dayconklt(arr))
```

```python
def swapnums(n):
    if 0 in n:
        pos = 0
        for i in range(-1, -len(n)-1, -1):
            if n[i] == 0:
                pos = i
                n[0], n[pos] = n[pos], n[0]
                return int("".join(n))
    else:
        for j in range(len(n)):
            min_num = n[-1]
            posk = -1
            check = False
            for k in range(-1, -len(n)+j, -1):
                if n[k] < n[j]:
                    check = True
                    if n[k] < min_num:
                        min_num = n[k]
                        posk = k
            if check:
                n[j], n[posk] = n[posk], n[j]
                return int("".join(n))
            if j == len(n)-1:
                n[-2], n[-1] = n[-1], n[-2]
                return int("".join(n))
n = int(input())
print(swapnums(list(str(n))))
```
