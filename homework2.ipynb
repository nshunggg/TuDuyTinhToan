Bai 3: Tinh tam giac

```python
import math
a, b, c = map(float, input().split())

if a + b> c and a + c > b and b + c > a:
    if a == b == c:
        print("Tam giac deu")
    elif a == b or a == c and b == c:
        print("Tam giac can")
    elif a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2:
        print("Tam giac vuong")
    else:
        print("Tam giac thuong")
    
    p = (a + b+ c)/2
    print("Dien tich hinh tam giac la :", math.sqrt(p*(p-a)*(p-b)*(p-c)))
    print ("Chu vi hinh tam giac la :", a + b + c)
else:
    print("Khong la ba canh cua mot tam giac")
```

Bai 5: Phuong trinh bac 2

```python
import math
a , b, c = map(int , input().split())

delta = b**2 -4*a*c
if delta < 0:
    print("Phuong trinh vo nghiem")
elif delta == 0:
    print("Phuong trinh co nghiem kep:", -b/(2*a))
else:
    x1 = (-b + math.sqrt(delta))/(2*a)
    x2 = (-b - math.sqrt(delta))/(2*a)
    print("Phuong trinh co hai nghiem phan biet:")
    print("x1 =", x1)
    print("x2 =", x2)
```

Bai 8: Giai he phuong trinh

Bai 10: Ktra xem thuoc duong tron
