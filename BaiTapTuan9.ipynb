```python
def merge(lst1, lst2, res = [], i = 0, j = 0):
    if i == len(lst1) or j == len(lst2):
        res.extend(lst1[i:])
        res.extend(lst2[j:])
        return res
    if lst1[i] < lst2[j]:
        res.append(lst1[i])
        i += 1
    else:
        res.append(lst2[j])
        j += 1
    return merge(lst1, lst2, res, i, j)
lst1 = list(map(int, input().split()))
lst2 = list(map(int, input().split()))
print(*merge(lst1, lst2))
```

```python
def kth_largest_in_sorted_matrix(mat, k):
    flat = []
    for row in mat:
        flat.extend(row)
    flat.sort()
    return flat[k-1] if 0 < k <= len(flat) else None
mat = []
for i in range(3):
    row = list(map(int, input().split()))
    mat.append(row)
k = int(input())
print(kth_largest_in_sorted_matrix(mat, k))
```

```python
def swap(a, i, pos):
    a[i], a[pos] = a[pos], a[i]
    return a
def swap_to_min_number(a, i = 0):
    min_i = a[i]
    pos = i
    for j in range(i+1, len(a)):
        if a[j] <= min_i:
            min_i = a[j]
            pos = j
    if i == len(a)-2:
        return swap(a, -1, -2)
    elif pos != i:
        return swap(a, i, pos)
    else:
        return swap_to_min_number(a, i+1)
n = int(input())
a = list(str(n))
print(int("".join(swap_to_min_number(a))))
```

```python
def trans_char(s1, s2, i = 0, j = 0):
    if i == len(s1):
        return s1
    if j == len(s2):
        return s1[:j]
    if s1[i] == s2[j]:
        return trans_char(s1, s2, i+1, j+1)
    elif chr(ord(s1[i])-32) == s2[j]:
        s1[i] = chr(ord(s1[i])-32)
        return trans_char(s1, s2, i+1, j+1)
    else:
        s1.pop(i)
        return trans_char(s1, s2, i, j)
n = int(input())
for _ in range(n):
    a = list(input())
    b = list(input())
    if trans_char(a, b) == b:
        print("YES")
    else:
        print("NO")
```

```python
def subset(lst1, lst2, i = 0, arr = []):
    n = len(lst1)
    if i == len(lst2) - n + 1:
        return arr
    if lst1[0] == lst2[i:i+n][0]:
        arr.append(lst2[i:i+n])
    return subset(lst1, lst2, i+1, arr)
lst1 = list(map(int, input().split()))
lst2 = list(map(int, input().split()))
if lst1 in subset(lst1, lst2):
    print("YES")
else:
    print("NO")
```

```python
def next_permutation(arr, n):
    i = n - 2
    while i >= 0 and arr[i] >= arr[i+1]:
        i -= 1
    if i == -1:
        return sorted(arr)
    j = n - 1
    while arr[j] <= arr[i]:
        j -= 1
    arr[i], arr[j] = arr[j], arr[i]
    l, r = i+1, n-1
    while l < r:
        arr[l], arr[r] = arr[r], arr[l]
        l += 1; r -= 1
    return arr
n = int(input())
arr = list(map(int, input().split()))
print(*next_permutation(arr, n))
```

```python
def canPlaceFlowers(flowerbed, k):
    n = len(flowerbed)
    i = 0
    while i < n and k > 0:
        if flowerbed[i] == 0:
            left_empty = (i == 0) or (flowerbed[i-1] == 0)
            right_empty = (i == n-1) or (flowerbed[i+1] == 0)
            if left_empty and right_empty:
                flowerbed[i] = 1
                k -= 1
                i += 2
                continue
        i += 1
    return k <= 0
flowerbed = list(map(int, input().split()))
k = int(input())
print(canPlaceFlowers(flowerbed, k))
```

```python
def is_path_crossing(moves):
    x = y = 0
    visited = {(0,0)}
    for ch in moves:
        if ch == 'R':
            x += 1
        elif ch == 'L':
            x -= 1
        elif ch == 'U':
            y += 1
        elif ch == 'D':
            y -= 1
        if (x,y) in visited:
            return True
        visited.add((x,y))
    return False
moves = input()
print(is_path_crossing(moves))
```

```python
def Median(nums):
    if len(nums) != 5:
        raise ValueError("Requires exactly 5 numbers")
    mn = min(nums)
    mx = max(nums)
    removed = False
    for i in range(len(nums)):
        if nums[i] == mn:
            nums.pop(i)
            removed = True
            break
    removed = False
    for i in range(len(nums)-1, -1, -1):
        if nums[i] == mx:
            nums.pop(i)
            removed = True
            break
    b0, b1, b2 = nums[0], nums[1], nums[2]
    return b0 + b1 + b2 - min(b0, b1, b2) - max(b0, b1, b2)
nums = list(map(int, input().split()))
print(Median(nums))
```

```python
def findRadius(houses, heaters):
    if not houses:
        return 0
    if not heaters:
        return None
    houses_sorted = sorted(houses)
    heaters_sorted = sorted(heaters)
    def nearest_distance(x):
        lo, hi = 0, len(heaters_sorted) - 1
        if x <= heaters_sorted[0]:
            return heaters_sorted[0] - x
        if x >= heaters_sorted[-1]:
            return x - heaters_sorted[-1]
        while lo <= hi:
            mid = (lo + hi) // 2
            if heaters_sorted[mid] == x:
                return 0
            if heaters_sorted[mid] < x:
                lo = mid + 1
            else:
                hi = mid - 1
        d1 = abs(heaters_sorted[lo] - x)
        d2 = abs(heaters_sorted[hi] - x)
        return d1 if d1 < d2 else d2
    radius = 0
    for h in houses_sorted:
        d = nearest_distance(h)
        if d is None:
            return None
        if d > radius:
            radius = d
    return radius
houses = list(map(int, input().split()))
heaters = list(map(int, input().split()))
print(findRadius(houses, heaters))
```
