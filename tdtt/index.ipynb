BIN SEARCH

```python
def binsearch(arr, target):
    n = len(arr)
    left , right = 0, n - 1
    while left <= right:
        mid = (left + right)//2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
arr = list(map(int, input().split()))
target = int(input())

print(binsearch(arr, target))
```

Bubble Sort

```python
def bubblesort(arr):
    n = len(arr)
    for i in range(n):
        swap = False
        for j in range(1, n-i):
            if arr[j] < arr[j-1]:
                arr[j], arr[j-1] = arr[j-1], arr[j]
                swap = True
    if not swap:
        return arr
arr = list(map(int, input().split()))
print(bubblesort(arr))
```

Selection Sort

```python
def selectionsort(arr):
    n = len(arr)
    for i in range(n):
        minvl = i
        for j in range(i+1, n):
            if arr[minvl] > arr[j]:
                minvl = j
        arr[i], arr[minvl] = arr[minvl], arr[i]
    return arr
arr = list(map(int, input().split()))
print(selectionsort(arr))
```

Insertion Sort

```python
def insertionsort(arr):
    n = len(arr)
    for i in range(1, n):
        a = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > a:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = a
    return arr
arr = list(map(int, input().split()))
print(insertionsort(arr))
```

Merge Sort

```python
# Tach mang ra cac mang nho
def mergesort(arr):
    n = len(arr)
    if n <= 1:
        return arr
    mid = n//2
    left, right = arr[:mid], arr[mid:]
    sortleft = mergesort(left)
    sortright = mergesort(right)
    return merge(sortleft, sortright)
# Ghep mang theo thu tu
def merge(l, r):
    newarr = []
    i = j = 0
    while i < len(l) and j < len(r):
        if l[i] < r[j]:
            newarr.append(l[i])
            i += 1
        else:
            newarr.append(r[j])
            j += 1
    newarr.extend(l[i:])
    newarr.extend(r[j:])
    return newarr

arr = list(map(int, input().split()))
print(mergesort(arr))
```

Quick Sort

```python
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr)-1]
    left = [i for i in arr if i < pivot]
    right = [j for j in arr if j > pivot]
    return quicksort(left) + [pivot] + quicksort(right)
arr = list(map(int, input().split()))
print(quicksort(arr))
```

Tháp Hà Nội

```python
count = 0
def thaphn(n, A, C, B):
    global count
    if n == 1:
        count += 1
        print(f"Chuyen 1 tu cot {A} sang cot {C} ")
        return
    thaphn(n-1, A, B, C)
    count += 1
    print(f"Chuyen {n} tu cot {A} sang cot {C}")
    thaphn(n-1, B, C, A)
n = int(input())
thaphn(n, "A", "C", "B")
print(count)
```

```python
def indent(level):
    return " "*(level)
def print_outline(outline, level):
    for i in outline:
        if isinstance(i, list):
            print_outline(i, level + 1)
        else:
            print(indent(level) + i)
Book = ['Preface', '1 Fundamentals', ['1.1 Basic Programming Model', '1.2 Data Abstraction', '1.3 Bags, Queues, and Stacks', ['Bags', 'Queues', 'Stacks'], '1.4 Analysis of Algorithms', '1.5 Case Study: Union-Find'], '2 Sorting', '3 Searching']
print_outline(Book, 0)
```

```python
def daonguoc(s, arr = [], i = 0):
    if i == len(s):
        return arr
    arr.insert(0, s[i])
    return daonguoc(s, arr, i+1)
s = input()
print("".join(daonguoc(s)))
```
