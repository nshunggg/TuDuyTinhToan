```python
# De Quy
def fib(n):
    if n <=1:
        return n
    else:
        return fib(n-1) + fib(n-2)

n = int(input())

print(fib(n))
```

```python
#Quy hoach dong
def fib(n):
    if n <= 1:
        return n
    else:
        dp = [0] * (n + 1)
        dp[0] = 0
        dp[1] = 1
        for i in range(2, n+1):
            dp[i] = dp[i-1] + dp[i-2]
        return dp[n]
    
n = int(input())
print(fib(n))
```
