Bài 3: Chuyển đổi Hoa - Thường

```python
#A-Z : 65-90
#a-z : 97-122
kytu = input("Nhập một ký tự:")
if 65<=ord(kytu)<=90:
    print(chr(ord(kytu)+32))
elif 97<=ord(kytu)<=122:
    print(chr(ord(kytu)-32))
else:
    print("INVALID")
```

Bài 6: Tính diện tích tam giác

```python
nhapso = input().split()
a, b, c = map(int, nhapso)
import math

if a+b>c or a+c>b or b+c>a:
    print("Đây là 3 cạnh của tam giác")
    p = (a+b+c)/2
    S = math.sqrt(p*(p-a)*(p-b)*(p-c))
    print(f"Diện tích của tam giác là : {S:.1f}")
else:
    print("Đây không phải 3 cạnh của tam giác")
```

Bài 8: Tính tiền điện

```python
tenchuho = input("Tên chủ hộ:")
chisothtrc = int(input("Chỉ số tháng trước:"))
chisothnay = int(input("Chỉ số tháng nay:"))

print("Họ và tên:", tenchuho)
sodien = chisothnay - chisothtrc
if 0<sodien<=50:
    tiendien = sodien*1984
elif 51<=sodien<=100:
    tiendien = 50*1984 + (sodien-50)*2050
elif 101<=sodien<=200:
    tiendien = 50*1984 + 50*2050 + (sodien-100)*2380
elif 201<=sodien<=300:
    tiendien = 50*1984 +50*2050 +100*2380 + (sodien-200)*2998
elif 301<=sodien<=400:
    tiendien = 50*1984 +50*2050 +100*2380 +100*2998 + (sodien-300)*3350
else:
    tiendien = 50*1984 +50*2050 +100*2380 +100*2998 +100*3350 + (sodien-400)*3460
print(f"Tiền phải trả là: {tiendien*1.08:.0f} đồng")
```

Bài 4: Kiểm tra tuổi đủ điều kiện bầu cử

```python
import datetime
n = int(input("Nhập năm sinh của bạn:"))

namht = datetime.datetime.today().year
if namht - n >=18:
    print("Bạn đủ tuổi để đi bầu cử")
else:
    print("Bạn chưa đủ tuổi để đi bầu cử")
```

Bài 5: So sánh 2 số

```python
numbers = input().split()
a, b = map(float, numbers)
if a > b:
    solonhon = a
    print(f"Số lớn hơn là: {solonhon}")
elif a< b:
    solonhon = b
    print(f"Số lớn hơn là: {solonhon}")
else:
    print("Hai số bằng nhau")
```
