W8A1: Xây dựng lớp Cylinder (hình trụ) gồm: bán kính và chiều cao

```python
class Cylinder:
    pi = 3.14
    def __init__(self, r, h):
        self.r = r
        self.h = h
    def areatp(self):
        return 2 * Cylinder.pi * self.r * self.h + 2 * Cylinder.pi * self.r ** 2
    def volume(self):
        return Cylinder.pi * self.r ** 2 * self.h
    def __str__(self):
        return f"Ban kinh: {self.r}, Chieu cao: {self.h}"
r, h = map(float, input().split())
c1 = Cylinder(r, h)
print(c1)
print(f"Dien tich xung quanh: {c1.areatp():.2f}")
print(f"The tich: {c1.volume():.2f}")
```

W8A2: Xây dựng lớp Date

```python
class MyDate:
    def __init__(self, d, m, y):
        self.d = d
        self.m = m
        self.y = y
    def invalid(self):
        maxday = -1
        match self.m:
            case 1|3|5|7|8|10|12:
                maxday = 31
            case 4|6|9|11:
                maxday = 30
            case 2:
                if (self.y % 4 == 0 and self.y % 100 != 0) or self.y % 400 == 0:
                    maxday = 29
                else:
                    maxday = 28
        return 1 <= self.d <= maxday and self.y > 0
    def nextday(self):
        self.d += 1
        nextd = MyDate(self.d, self.m, self.y)
        if nextd.invalid():
            return nextd
        self.d = 1
        self.m += 1
        if nextd.invalid():
            return nextd
        self.m = 1
        self.y += 1
        return nextd
    def __str__(self):
        return f"{self.d:02d}/{self.m:02d}/{self.y:02d}"
d, m, y = map(int, input().split("/"))
date = MyDate(d, m, y)
if not date.invalid():
    print("INVALID")
else:
    date.nextday()
    print(date)
```

W8A3: Xây dựng lớp điểm với hai thuộc tính x, y

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def toado(self):
        if self.x == 0 and self.y == 0:
            return "M la goc toa do O(0,0)"
        elif self.x == 0 and self.y != 0:
            return "M nam tren truc tung"
        elif self.x != 0 and self.y == 0:
            return "M nam tren truc hoanh"
        else:
            return "M khong phai goc toa do va khong nam tren truc toa do"
    def khoangcach(self, other):
        return ((self.x - other.x)**2 + (self.y - other.y)**2)**0.5
x, y = map(int, input().split())
M = Point(x, y)
O = Point(0, 0)
print(M.toado())
print(f"{M.khoangcach(O):.2f}")
    
```

W8A5: Xây dựng lớp ShoppingCart

```python
class ShoppingCart:
    def __init__(self):
        self.lst = []
    def add_item(self, name, price):
        item = {"name": name, "price": price}
        self.lst.append(item)
    def remove_item(self, name):
        for i in self.lst:
            if i["name"] == name:
                self.lst.remove(i)
                return
        print("Khong tim thay san pham")
    def kiem_tra(self):
        return len(self.lst) == 0
    def total_price(self):
        tong = 0
        for item in self.lst:
            tong += item["price"]
        return tong
    def hide_item(self):
        for item in self.lst:
            print(f"Ten: {item["name"]} - Gia: {item["price"]}")
    def clear_cart(self):
        self.lst.clear()
        return "Gio hang da duoc lam rong"
cart = ShoppingCart()
while True:
    op = input()
    if op == "2":
        name = input()
        price = float(input())
        cart.add_item(name, price)
    elif op == "3":
        name = input()
        cart.remove_item(name)
    elif op == "4":
        if cart.kiem_tra():
            print("Gio hang trong")
        else:
            print("Gio hang khong trong")
    elif op == "5":
        print(f"Tong gia tri: {cart.total_price():.2f}")
    elif op == "6":
        cart.hide_item()
    elif op == "7":
        print(cart.clear_cart())
    else:
        print("Ban co muon thoat khoi chuong trinh khong?")
        thoat = input()
        if thoat in ["yes", "YES", "Yes"]:
            print("Tam biet, hen gap lai!")
            break
```

W8A6: Xây dựng lớp Rectangle(w, h)

```python
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h
    def area(self):
        return self.w * self.h
    def perimeter(self):
        return 2*(self.w + self.h)
    def scale(self, k):
        self.w *= k
        self.h *= k
        return Rectangle(self.w, self.h)
w, h, k = map(float, input().split())
hcn = Rectangle(w, h)
hcn.scale(k)
print(f"Dien tinh : {hcn.area():.2f}")
print(f"Chu vi : {hcn.perimeter():.2f}")
```

W8A7: Lớp Fraction(num, den)

```python
class Fraction:
    def __init__(self, num, den):
        gcd = 1
        for i in range(2, min(abs(num), abs(den))+1):
            if num % i == 0 and den % i == 0:
                if i > gcd:
                    gcd = i
        num //= gcd
        den //= gcd
        if den < 0:
            den = -den
            num = -num
        self.n = num
        self.d = den
    def __add__(self, other):
        if self.d == other.d:
            self.n += other.n
        else:
            self.n = self.n * other.d + other.n * self.d
            self.d *= other.d
        return Fraction(self.n, self.d)
    def __sub__(self, other):
        if self.d == other.d:
            self.n -= other.n
        else:
            self.n = self.n * other.d - other.n * self.d
            self.d *= other.d
        return Fraction(self.n, self.d)
    def __mul__(self, other):
        self.n *= other.n
        self.d *= other.d
        return Fraction(self.n, self.d)
    def __truediv__(self, other):
        if other.n == 0:
            return Fraction(0, self.d)
        else:
            self.n *= other.d
            self.d *= other.n
            return Fraction(self.n, self.d)
    def __str__(self):
        if self.n == 0:
            return "0"
        return f"{self.n}/{self.d}"
a, b, op, c, d = map(str, input().split())
num1 = Fraction(int(a), int(b))
num2 = Fraction(int(c), int(d))
if op == "+":
    num3 = num1 + num2
    print(num3)
elif op == "-":
    num3 = num1 - num2
    print(num3)
elif op == "*":
    num3 = num1*num2
    print(num3)
elif op == "/":
    num3 = num1 / num2
    print(num3)


        
```

W8A8: Lớp BankAccount(owner, balance)

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.o = owner
        self.b = balance
    def deposit(self, x):
        self.b += x
        return self.b
    def withdraw(self, x):
        if self.b > x:
            self.b -= x
        return self.b
    def __str__(self):
        return f"{self.o}: {self.b:.2f}"
owner, balance = map(str, input().split())
account1 = BankAccount(owner, int(balance))
n = int(input())
for _ in range(n):
    nap_rut, x = map(str, input().split())
    if nap_rut == "DEPOSIT":
        account1.deposit(float(x))
    elif nap_rut == "WITHDRAW":
        account1.withdraw(float(x))
print(account1)
```

```python
class Student:
    def __init__(self, name):
        self.n = name
        self.sub = []
    def add_score(self, subject, score):
        m = {"Subject": subject, "Score": score}
        self.sub.append(m)
        return
    def avg(self):
        tong = 0
        for sub in self.sub:
            tong += sub["Score"]
        return tong/len(self.sub)
    def rank(self):
        if self.avg() >= 8:
            return "Excellent"
        elif 6.5 <= self.avg() < 8:
            return "Good"
        elif 5 <= self.avg() < 6.5:
            return "Average"
        else:
            return "Poor"
    def __str__(self):
        return f"{self.n} {self.avg():.2f} {self.rank()}"
name = input()
student = Student(name)
n = int(input())
for _ in range(n):
    subject, score = map(str, input().split())
    student.add_score(subject, float(score))
print(student)
```

```python
import math
class Points:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
    # Tim toa do vecto
    def __sub__(self, other):
        return Points(self.x - other.x, self.y - other.y, self.z - other.z)
    # Tich vo huong
    def tichVH(self, other):
        return self.x*other.x + self.y*other.y + self.z*other.z
    def tichCH(self, other):
        return Points(self.y*other.z-self.z*other.y, self.z*other.x-self.x*other.z, self.x*other.y-self.y*other.x)
    def dodai(self):
        return (self.x**2 + self.y**2 + self.z**2)**0.5
points = []
for _ in range(4):
    x, y, z = map(float, input().split())
    point = Points(x, y, z)
    points.append(point)
A, B, C, D = points[0], points[1], points[2], points[3]
nabc = (B-A).tichCH(C-A)
nbcd = (C-B).tichCH(D-B)
cosphi = nabc.tichVH(nbcd)/(nabc.dodai()*nbcd.dodai())
do = math.degrees(math.acos(cosphi))
print(f"{do:.2f} *")
```

W8A12: ComplexNumber

```python
class ComplexNumber:
    def __init__(self, real, ima):
        self.r = real
        self.i = ima
    def multiply(self, other):
        return ComplexNumber(self.r * other.r - self.i * other.i, self.i * other.r + self.r * other.i)
    def __str__(self):
        return f"{self.r} + {self.i}i"
r, i = map(int, input().split())
r1, i1 = map(int, input().split())
num1 = ComplexNumber(r, i)
num2 = ComplexNumber(r1, i1)
print(num1)
print(num2)
print(num1.multiply(num2))
```

```python
class Triangle:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    def getArea(self):
        if self.a + self.b <= self.c or self.b + self.c <= self.a or self.a + self.c <= self.b:
            return -1
        p = (self.a + self.b + self.c)/2
        return (p*(p-self.a)*(p-self.b)*(p-self.c))**0.5
    def __str__(self):
        if self.getArea() == -1:
            return "Invalid"
        return f"Dien tich: {self.getArea():.2f}"
a, b, c = map(int, input().split())
triangle = Triangle(a, b, c)
print(triangle)
```
