```python
def binnary(n, m, r = ""):
    if n == 0 and m == 0:
        print(r)
        return
    if n > 0:
        binnary(n-1, m, r + "0")
    if m > 0:
        binnary(n, m-1, r + "1")
n, m = map(int, input().split())
binnary(n, m)
```

```python
def daycontang(a):
    if not a: return 0
    n = len(a)
    dp = [1]*n
    best = 1
    for i in range(1, n):
        for j in range(i):
            if a[j] < a[i] and dp[j]+1 > dp[i]:
                dp[i] = dp[j]+1
        if dp[i] > best: best = dp[i]
    return best
a = list(map(int, input().split()))
print(daycontang(a))
```
