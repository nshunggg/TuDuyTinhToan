```python
def bin_search(arr, target):
    left, right = 0, len(arr)-1
    while left <= right:
        mid = (left + right)//2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
arr = list(map(int, input().split()))
target = int(input())
print(bin_search(arr, target))
```

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swap = False
        for j in range(1, n-i):
            if arr[j-1] > arr[j]:
                arr[j-1], arr[j] = arr[j], arr[j-1]
                swap = True
    if not swap:
        return arr
    
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_value = i
        for j in range(i+1, n):
            if arr[j] < arr[min_value]:
                min_value = j
        arr[i], arr[min_value] = arr[min_value], arr[i]
    return arr
arr = list(map(int, input().split()))
print(bubble_sort(arr))
print(selection_sort(arr))
```

```python
def time(interval, q):
    listi = []
    for i in interval:
        if i[0] <= q <= i[1]:
            listi.append((i[1]-i[0] + 1))
    if len(listi) == 0:
        return -1
    dodai = min(listi)
    return dodai
    
interval = eval(input())
queries = eval(input())
out = []
for q in queries:
    out.append(time(interval, q))
print(out)
```
