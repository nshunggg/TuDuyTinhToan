```python
def nguyento(n):
    ktra = True
    for i in range(2, n):
        if n % i == 0:
            ktra = False
    return ktra
n = int(input())
print(nguyento(n))
```

W5A4

```python
def perfect(n):
    tonguoc = 0
    for i in range(1, n):
        if n % i == 0:
            tonguoc += i
    if tonguoc == n:
        return "YES"
    else:
        return "NO"
n = int(input())

print(perfect(n))
```

W5A5

```python
def find(n, k):
    for i in range(len(n)):
        if n[i] == k:
            kq = i
            break
        else:
            kq = -1
    return kq
n = list(map(int, input().split()))
k = int(input())
print(find(n, k))
```

W5A6

```python
def giaithua(n):
    gthua = 1
    i = 1
    while i <= n:
        gthua *= i
        i += 1
    return gthua
n = int(input())
print(giaithua(n))
```

W5A7

```python
def maytinh(a, b, n):
    match n:
        case "-":
            return a - b
        case "+":
            return a + b
        case "/":
            return a / b
        case "*":
            return a * b

a = float(input())
n = input()
b = float(input())
print(f"{a} {n} {b} = {maytinh(a, b, n):.2f}")
```

W5A10

```python
def dangcau(s, t):
    if len(s) != len(t):
        return False
    return len(set(s)) == len(set(t)) == len(set(zip(s, t)))

s, t = map(str, input().split())

print(dangcau(s, t))
```

```python
def triangle(a, b, c):
    if a + b > c and a + c > b and b + c > a:
        return a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2

a, b, c = map(int, input().split())

print(triangle(a, b, c))
```
