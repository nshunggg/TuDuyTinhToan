## 1. Class và Object là gì?
- **Class**: bản thiết kế (lớp các đối tượng nào đó)
- **Object**: đối tượng được tạo từ class (lớp)


```python

class Person:
    def __str__ (self):
        return "Hello"

p1 = Person() # gọi __init__ mặc định của python - KHông làm gì cả, chỉ cho phép tạo đối tượng mà
              #không tạo thêm thuộc tính nào
print(p1)
```


## 2. Thuộc tính (Attributes) và phương thức (Methods)


```python

class Person:
    def __init__(self, n, a):
        self.name = n #sinh ra thuộc tính tên là name
        self.age = a #tạo ra thuộc tính tên là age

    def greet(self):
        return f"Xin chao, toi la {self.name}, {self.age} tuoi"

p = Person("Long", 20)
print(p.greet())
```


## 3. self là gì?
- `self` đại diện cho **chính object đang gọi**


```python

class Counter:
    def __init__(self):
        self.value = 0 #value là tên thuộc tính của lớp Counter

    def increase(self): #increase là tên phương thức của lớp Counter
        self.value += 1

c = Counter() #Dùng hàm tại __init__ de tạo ra đối tượng c
c.increase() # Gọi tới phương thức increase từ đối tượng c
c.increase()
print(c.value)
```


## 4. Class variable (Biến của lớp) và Instance variable(Biến của đối tượng)


1. Instance variable (Biến của đối tượng)
 **👉**: Là biến thuộc về TỪNG object riêng lẻ
    - Được khai báo thông qua self
    - Mỗi object có bản sao riêng
    - Thay đổi ở object này KHÔNG ảnh hưởng object khác

2. Class variable (Biến của lớp)
👉 Là biến dùng CHUNG cho tất cả các object thuộc lớp
   - Khai báo trực tiếp trong class
   - Chỉ có 1 bản duy nhất
   - Mọi object đều truy cập chung


```python

class Student:
    school = "PTIT"   # class variable

    def __init__(self, name):
        self.name = name  # instance variable

s1 = Student("A")
s2 = Student("B")

print(s1.school, s2.school)
Student.school = "VNU" # Thay đổi giá trị của class variable

print(s1.school, s2.school)

print(s1.name, s2.name)
s1.name = "C"
print(s1.name, s2.name)
```


## 5. Các phương thức đặc biệt (__dunder__ methods)
#### Đọc thêm bên file kia :D 


```python

class Point:
    def __init__(self, _x, _y): # hàm - phương thức khởi tạo
        self.x = _x
        self.y = _y

    def __str__(self): #hàm - phương thức in (print) đối tượng
        return f"({self.x}, {self.y})"

    def __add__(self, other):# Hàm - phương thức cộng đối tượng với 1 đối tượng khác
        return Point(self.x + other.x, self.y + other.y)
    
    def __del__ (self):#Hàm - Phương thức hủy đối tượng - Tự động được gọi khi kết thúc
        print("Object deleted") # 

p1 = Point(1, 2)
p2 = Point(3, 4)
print(p1)
print(p1 + p2)
```


## 6. Kế thừa (Inheritance)


```python

class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

d = Dog()
d.speak()
```


## 7. super()


```python

class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name) #gọi hàm tạo của lớp Animal
        self.breed = breed

dog = Dog("Lucky", "Husky")
print(dog.name, dog.breed)
```


## 8. Encapsulation – Đóng gói (_ và __)


```python

class Account:
    def __init__(self, balance):
        self._balance = balance

    def get_balance(self):
        return self._balance

acc = Account(1000)
print(acc.get_balance())
```


## 9. Property (@property)


```python

class Circle:
    def __init__(self, r):
        self._r = r

    @property
    def area(self):
        return 3.14 * self._r ** 2

c = Circle(5)
print(c.area)
```


## 10. Classmethod & Staticmethod


```python

class MathUtils:
    pi = 3.14

    @classmethod
    def circle_area(cls, r):
        return cls.pi * r * r

    @staticmethod
    def add(a, b):
        return a + b

print(MathUtils.circle_area(3))
print(MathUtils.add(2, 5))
```


## 12. Ví dụ tổng hợp (Mini Project)


```python

class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
        else:
            print("Không đủ tiền")

    def __str__(self):
        return f"{self.owner}: {self.balance}"

acc = BankAccount("Long", 1000)
acc.deposit(500)
acc.withdraw(300)
print(acc)
```
