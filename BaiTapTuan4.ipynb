```python
while True:
    Si = 0
    a, b = map(int, input().split())
    if a <= b:
        for i in range(a, b+1):
            ktra = True
            for j in range(2, i):
                if i % j == 0:
                    ktra = False
            if i == 1:
                ktra = False
            if ktra:
                Si += i
        print(f"Tong cac so nguyen to trong doan [{a}, {b}] la: {Si}")
    else:
        break
```

```python
n = int(input())

max_dodai = 0
best_x = 0
for x in range(1, n+1):
    dodai = 0
    a = x
    while True:
        dodai += 1
        if x == 1:
            break
        if x % 2 == 0:
            x *= 0.5
        else:
            x = 3*x + 1
    if dodai > max_dodai:
        max_dodai = dodai
        best_x = a
print(best_x, max_dodai)
```

W4A14

```python
m, n = map(int, input().split())

gcd = 0
for i in range(1, min(m, n)+1):
    if m % i == 0 and n % i == 0:
        if i > gcd:
            gcd = i
print(f"Uoc chung lon nhat cua {m} va {n} la : {gcd}")
```
